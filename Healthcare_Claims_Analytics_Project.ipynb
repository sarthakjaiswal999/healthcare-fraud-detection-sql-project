```python
con.execute("""

WITH monthly_fraud AS (

    SELECT

        DATE_TRUNC(
            'month',
            CAST(Claim_Date AS DATE)
        ) AS claim_month,

        SUM(
            CASE
                WHEN Is_Fraudulent = TRUE THEN 1
                ELSE 0
            END
        ) AS fraudulent_claims

    FROM claims

    GROUP BY claim_month

),

fraud_trend AS (

    SELECT

        claim_month,

        fraudulent_claims,

        LAG(fraudulent_claims) OVER(
            ORDER BY claim_month
        ) AS previous_month_fraud

    FROM monthly_fraud

)

SELECT

    claim_month,

    fraudulent_claims,

    previous_month_fraud,

    fraudulent_claims - previous_month_fraud
        AS fraud_change,

    ROUND(
        (
            (fraudulent_claims - previous_month_fraud)
            * 100.0
        ) / previous_month_fraud,
        2
    ) AS fraud_change_percentage

FROM fraud_trend

ORDER BY claim_month

""").fetchdf()
```

```python
con.execute("""

WITH risk_scoring AS (

    SELECT

        Claim_ID,

        (

            CASE
                WHEN Claim_Submitted_Late = TRUE
                THEN 25
                ELSE 0
            END

            +

            CASE
                WHEN Length_of_Stay_Days <= 3
                THEN 20
                ELSE 0
            END

            +

            CASE
                WHEN Number_of_Previous_Claims_Patient >= 8
                THEN 15
                ELSE 0
            END

            +

            CASE
                WHEN Claim_Amount >= 1500000
                THEN 30
                ELSE 0
            END

        ) AS risk_score,

        Is_Fraudulent

    FROM claims

)

SELECT

    CASE

        WHEN risk_score >= 60
        THEN 'High Risk'

        WHEN risk_score >= 30
        THEN 'Medium Risk'

        ELSE 'Low Risk'

    END AS risk_segment,

    COUNT(*) AS total_claims,

    SUM(
        CASE
            WHEN Is_Fraudulent = TRUE
            THEN 1
            ELSE 0
        END
    ) AS fraudulent_claims,

    ROUND(

        SUM(
            CASE
                WHEN Is_Fraudulent = TRUE
                THEN 1
                ELSE 0
            END
        ) * 100.0 / COUNT(*),

        2

    ) AS fraud_percentage

FROM risk_scoring

GROUP BY risk_segment

ORDER BY fraud_percentage DESC

""").fetchdf()
```

```python
con.execute("""

WITH claim_percentiles AS (

    SELECT

        Claim_ID,

        Provider_Specialty,

        Claim_Amount,

        Is_Fraudulent,

        NTILE(10) OVER(
            ORDER BY Claim_Amount DESC
        ) AS claim_decile

    FROM claims

)

SELECT

    claim_decile,

    COUNT(*) AS total_claims,

    ROUND(
        AVG(Claim_Amount),
        2
    ) AS avg_claim_amount,

    SUM(
        CASE
            WHEN Is_Fraudulent = TRUE
            THEN 1
            ELSE 0
        END
    ) AS fraudulent_claims,

    ROUND(

        SUM(
            CASE
                WHEN Is_Fraudulent = TRUE
                THEN 1
                ELSE 0
            END
        ) * 100.0 / COUNT(*),

        2

    ) AS fraud_percentage

FROM claim_percentiles

GROUP BY claim_decile

ORDER BY claim_decile

""").fetchdf()
```

```python
con.execute("""

WITH yearly_cohorts AS (

    SELECT

        EXTRACT(
            YEAR FROM CAST(Claim_Date AS DATE)
        ) AS claim_year,

        Claim_Amount,

        Is_Fraudulent

    FROM claims

)

SELECT

    claim_year,

    COUNT(*) AS total_claims,

    ROUND(
        AVG(Claim_Amount),
        2
    ) AS avg_claim_amount,

    SUM(
        CASE
            WHEN Is_Fraudulent = TRUE
            THEN 1
            ELSE 0
        END
    ) AS fraudulent_claims,

    ROUND(

        SUM(
            CASE
                WHEN Is_Fraudulent = TRUE
                THEN 1
                ELSE 0
            END
        ) * 100.0 / COUNT(*),

        2

    ) AS fraud_percentage

FROM yearly_cohorts

GROUP BY claim_year

ORDER BY claim_year

""").fetchdf()
```
